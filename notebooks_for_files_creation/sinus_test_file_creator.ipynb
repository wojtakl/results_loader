```python
import numpy as np
import matplotlib.pyplot as plt
import json
```

```python
for i in range(10):
    time = np.linspace(0, 5000, 5000) # 5s in ms
    freq = np.random.randint(1,4)
    sin = np.sin(2*np.pi*freq*time/1000) # 1 hz
    is_valid = np.random.randn() > 0.5
    noise_multiplyer = 1 if is_valid else 3
    sin_with_noise = sin + (np.random.randn(5000)*0.1*noise_multiplyer)
    order_number = np.random.randint(900000,999999)

    context_info = {"order_number": order_number,
                    "frequency": freq,
                    "time": 5000}
    
    with open(f"SinusTest_{order_number}.txt", "w") as file:
        file.write(json.dumps(context_info)+";"+",".join([str(t) for t in time])+";"+",".join([str(t) for t in sin_with_noise]))   
```
